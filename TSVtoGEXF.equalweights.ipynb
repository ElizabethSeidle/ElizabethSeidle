```python
import json, ast, math, time, collections
import networkx as nx
import codecs, csv, re

nodeIn = "Nodes.txt"
linkIn = "Edges.txt"
fileout = "PPPID_gephi_file.gexf"

f = open(nodeIn, 'rt')
nodes = []
nodesAttributes = {}
try:
    reader = csv.reader(f, delimiter='\t')
    header = reader.next()
    body = [row for row in reader]

finally:
    f.close()

for row in body:
    name = row[0].strip()
    nodes.append(name)
    rank = row[1].strip()
    nodesAttributes[name] = rank


f = open(linkIn, 'rt')
links = []
linksAttArray = []
linkAttributes = {}
try:
    reader = csv.reader(f, delimiter='\t')
    header = reader.next()
    body = [row for row in reader]

finally:
    f.close()

for row in body:
    s = row[0].strip()
    t = row[1].strip()
    if (t.lower()<s.lower()): 
        temp = s
        s = t
        t = temp
    linksAttArray.append([s,t,row[4]])
    if [s,t,row[4]] not in links:
        if [t,s,row[4]] not in links:
            links.append([s,t,row[4]])

for l in linksAttArray:
    linkAttributes[(l[0],l[1],l[2])] = 1
#for l in linksAttArray:
#    linkAttributes[(l[0],l[1],l[2])] = linkAttributes[(l[0],l[1],l[2])] + 1
#print linkAttributes
    
x = ""

x = x + """<?xml version="1.0" encoding="UTF-8"?>\n"""
x = x + """<gexf xmlns="http://www.gexf.net/1.2draft" version="1.2">\n"""
x = x + """<meta lastmodifieddate="2016-02-03">\n"""
x = x + """<creator>Gexf.net</creator>\n"""
x = x + """</meta>\n"""
x = x + """<graph mode="static" defaultedgetype="undirected">\n"""

x = x + """<attributes class="node">\n"""
x = x + """<attribute id="0" title="facRank" type="string"/>\n"""
x = x + """</attributes>\n"""
x = x + """<nodes>\n"""

for n in nodes:
    rn = ""
    rn = rn + """<node id="%s" label="%s">\n""" % (nodes.index(n), n.decode("utf-8"))
    rn = rn + """<attvalues>\n"""
    rn = rn + """<attvalue for="0" value="%s"/>\n""" % (nodesAttributes[n])
    rn = rn + """</attvalues>\n"""
    rn = rn + """</node>\n"""
    x = x + rn
x = x + """</nodes>\n"""
x = x + """<edges>\n"""
c = 0
for l in links:
    rl = """<edge id="%s" source="%s" target="%s" kind="%s" weight="%s"/>\n""" % (c, nodes.index(l[0]), nodes.index(l[1]), l[2],linkAttributes[(l[0],l[1],l[2])])
    x = x + rl
    c = c + 1
x = x + """</edges>\n"""


x = x + """</graph>\n"""
x = x + """</gexf>\n"""

f = codecs.open(fileout, "wb", "utf-8") 
f.write(x)
f.close()
```
